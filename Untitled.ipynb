```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
```

```python
######################################
###Reflect Tables into SQLAlchemy ORM
#####################################
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
###########################################
# Database Setup
###########################################
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
####################################
##Exploratory Climate Analysis
#####################################
```

```python
# Calculate the date one year from the last date in data set.
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)
```

```python
###################################
## Retrieve Precipitation Scores
###################################
```

```python
results = session.query(Measurement.date, Measurement.prcp)
# print(results.all())
```

```python
#############################################################
##Filter out all irrelevant information(Previous years data)
#############################################################
```

```python
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
# print(results)
```

```python
###############################################################################################################
## Save to the dataframe.... Use our results as one paramneter and specify columns names as our second parameter#
###############################################################################################################
df = pd.DataFrame(results, columns=['date','precipitation'])
```

```python
#############################################################################################################################
#### Use the set_index() Function  (use Inplace to determine if we want to create a new Dataframe) (inplace=True no new DF!!)
##############################################################################################################################
df.set_index(df['date'], inplace=True)

## Print the DF with the index
# print(df)
```

```python
####### With the index##########
# print(df)
```

```python
#########################################
##Sort the Dataframe
#########################################
```

```python
## Sort values by date using sort_index() function to sort with index
df = df.sort_index()
```

```python
### Sort values without index... set index=False
print(df.to_string(index=False))
```

```python
df.plot()
###Clean Up the Plot###
plt.xticks(rotation=90)
plt.ylabel("Inches", fontsize = 15)
plt.xlabel("Date", fontsize = 15)
plt.title("Preipitation Analysis", fontsize=18)

##### Save the figure
plt.savefig("Precipitation Plot.jpg")
```

```python
####################################
##Find the number of stations
###################################
```

```python
###### Count the number of stations using count() - Reference Station.station + .all to return results as a list
session.query(func.count(Station.station)).all()
```

```python
######## Determine the most active stations using groupby() - add the order_by function - add the all() function
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
##############################################
## Find the low, High, and Average Temperatures
###############################################
```

```python
### Use the results for the active station (USC00519281, 2772) to find the min, max, and avg using func.min, func.max etc
### Filter out all other stations + all to return results as a list (Total Observations - TOBS)
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).filter(Measurement.station == 'USC00519281').all()
```

```python
#################################################
## Plot the highest number of observations
################################################
```

```python
## Create a Query for the Temperature Observations 
## filter out the stations except for the most active one 
## filter for the most recent year  
## add all() to return results as a list
results = session.query(Measurement.tobs).\
filter(Measurement.station == 'USC00519281').\
filter(Measurement.date >= prev_year).all()

print(results)
```

```python
### Put the list into a Dataframe
df = pd.DataFrame(results, columns=['tobs'])
# Print with the index
# print(df)

#3  Print without the index
print(df.to_string(index=False))
```
